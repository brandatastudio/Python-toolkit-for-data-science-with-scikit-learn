### Gather :
when we have two colums that contain exactly the same type and feature data, we can create a colum
that represents that data and another one that represents categorical information, related to where each data came from.

```python
import pandas 
names = [ 'Wilbur', 'Petunia', 'Gregory' ]
a = [ 67, 80, 64 ]
b = [ 56, 90, 50 ]
df = pandas.DataFrame({'names':names,'a':a,'b':b})

# Here we have a dataframe with the colums a,b both of them contain the same type of data, the heart rate, but we have created
# two entire features to represent that data. This is only benefitial if we are doing a logistic regression on them.
# We can use gather to solve this. Transforming the colum names into data and joining both colums a and b. 

df
```

```python
def gather( df, key, value, cols ):
    
    "Arguments df represent the dataframe, key ill represent the name of the colum that will contain \
    the originals dataframes colum names, value will represent the name of the colum that will contain \
    the originals row data that was separated in different colums before, finally cols will be a list with the \
    names of the dataframes colums that will be affected. "
    
    id_vars = [ col for col in df.columns if col not in cols ]
    id_values = cols
    var_name = key
    value_name = value
    return pandas.melt( df, id_vars, id_values, var_name, value_name )

gather( df, 'drug', 'heartrate', ['a','b'] )
```

### Separate :
now other times, we have a column of strings that includes too much information, and that can be easily separated by a particualr element. For example the nex dataframe, that will include an aliment and the sauce we eat it with. The dataframe
would be cleaner with each string in it´s own column, representing an unique variable.

```python
messy = pd.DataFrame({ 
                      'food and sauce' : ["cucumber,tomatosauce", "apple,mustard", "pineapple, mayonaise"],
                      'b' : [3, 5, 6],
                      'c' : [7, 8, 9]})
```

```python
def separatestringcolum(df,originalcolumn,newcolum1,newcolum2,separator):
    
    "this function, on the other hand, will separate a column which elements \
    are strings , into two new colums with those strings. The arguments are \
    df as the dataframe, originalcolumn, being the column that we are going to split\
    newcolum1 is the name of the new created column that will have elements to the left\
    of the separator , and newcolumn2 the name of the second column, including the elements to \
    the right of the separator. Separator is a string specifying the element we will use to\
    separate"
    
    
    df[newcolum1], df[newcolum2] = df[originalcolumn].str.split(separator, 1).str
    
    return(df)


separatestringcolum(messy,"food and sauce", "food", "sauce" , ",")
    
```

### Spread :
The function spread, allows to distribute all the data in one colum of the dataframe, into new colums based of categories of another
categorical colum inside the dataframe.  For example the next dataframe, we will distribute "fc" data into many new colums, one 
for each category available in "sample_id". It has the opposite effect of gather. 

```python
import pandas as pd
import numpy as np
df = pd.DataFrame({
               'fc': [100,100,112,1.3,14,125],
               'sample_id': ['S1','S1','S1','S2','S2','S2'],
               'gene_symbol': ['a', 'b', 'c', 'a', 'b', 'c'],
               })

df = df[['gene_symbol', 'sample_id', 'fc']]
df
```

```python
p = pd.crosstab(index=df.gene_symbol, columns=df.sample_id, 
                     values=df.fc, aggfunc='mean') \
.rename_axis(None,1) \
.reset_index()



p
```

### Uniting two colums 

As simple as that, creating one colum that includes data from other two. 

#### Both colums are type string

```python
import pandas as pd
df = pd.DataFrame({'Quarter':'q1 q2 q3 q4'.split(), 'Year':'2000'})
df
```

```python
df['Period'] = df['Year'] + ' ' + df['Quarter']

df

type(df["Period"][1])
```

#### String colum + number colum

```python
import pandas as pd
df = pd.DataFrame({'Quarter':'q1 q2 q3 q4'.split(), 'Year':2000})
df
```

```python
df['Period'] = df['Year'].astype(str) + ' ' + df['Quarter']

df
```

#### Both colums are numbers

```python
import pandas as pd
df = pd.DataFrame({'Quarter': [1,2,3,4], 'Year': 2000})
df
```

```python
df['Period'] = df['Year'] * df['Quarter']

df
```
