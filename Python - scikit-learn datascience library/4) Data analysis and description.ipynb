```python
# This function transforms a cateogiral variable, into a numerical one. 
def categoricaltonum(dataset,category,mapdictionary):
    'this function transforms categorial variable of a dataset into numerical one, we specify as \
    arguments the dataset, the category we want to transform, and we use a dictionary to specify\
    which category will correspond to what created number afterwards'
    
    dataset[category] = dataset[category].map(mapdictionary)
    return(dataset)

# this is equivalent to : winedata['color'] = winedata['color'].map({'red': 1, 'white': 0})


categoricaltonum(winedata,"color",{'red': 1, 'white': 0})
```

```python
import pandas as pd

winexsc = pd.read_csv('C:\\Users\\stefano\\Documents\\datahack\\modulo python\\Ejercicios\\winequality.csv',sep = ';')

def allbasicstats(dataframeargument):
    import numpy as np 
    from scipy import stats
    import pandas as pd
    
    " Like describe method, but calculats also the median and the mode, and the outliers for each variable, you need to make sure \
    that all variables are numeric, only including those variables in the dataframe that are numeric "
    
    totallist = []
    for i in dataframeargument:
        ex = np.array(dataframeargument[i])
        exmode = stats.mode(ex)
        mean = ex.mean()
        mini = ex.min()
        maxi = ex.max()
        vari = ex.var()
        tipicd = ex.std() 
        mediana = np.median(ex)
        quantilearr = np.quantile(ex,np.array([0,0.25,0.50,0.75,1]))
        interquantile_range = quantilearr[3].astype(float) - quantilearr[1].astype(float)
        lower_inner_fence = quantilearr[1].astype(float) - (1.5 * interquantile_range)
        upper_inner_fence = quantilearr[3].astype(float) + (1.5 * interquantile_range)
        lower_outer_fence = quantilearr[1].astype(float) - (3 * interquantile_range)
        upper_outer_fence = quantilearr[3].astype(float) + (3 * interquantile_range)
        
       

        dictionaryresult = {"name" : str(i) , "mean" : mean, "minimal" : mini , "Maximal": maxi ,
                            "Var" : vari, "Std" : tipicd, "median" : mediana, "mode" : exmode , "Upper_out_fence" : upper_outer_fence , 
                            "Lower_out_fence" : lower_outer_fence , 
                             "lower_in_fence" : lower_inner_fence , "up_in_fence" :  upper_inner_fence , "quantiles" : quantilearr }
        totallist.append(dictionaryresult)
        
    totallist = pd.DataFrame(totallist)
    totallist.set_index("name", inplace = True)
    totallisttrans = totallist.T
    return(totallisttrans)
        
allbasicstats(winedata)
```

```python
def allbasicstatandcofs(dataframeargument):
    import numpy as np 
    from scipy import stats
    import pandas as pd
    from scipy import stats
    
    " Calculates all main stats for each of the variables of the data set, and three coeficients to individually \
    analyze them , pearson variation coefficient, kurtosis coefficient, and fisher's skewness. In the statistics \
    department, we calculate the mean, the median, the mode ( the value and its frequency ) , the quantiles, and the \
    outliers both inner and outer, for both upper and lower fence"
    
    totallist = []
    for i in dataframeargument:
        ex = np.array(dataframeargument[i])
        exmode = stats.mode(ex)
        mean = ex.mean()
        mini = ex.min()
        maxi = ex.max()
        vari = ex.var()
        tipicd = ex.std() 
        mediana = np.median(ex)
        quantilearr = np.quantile(ex,np.array([0,0.25,0.50,0.75,1]))
        interquantile_range = quantilearr[3].astype(float) - quantilearr[1].astype(float)
        lower_inner_fence = quantilearr[1].astype(float) - (1.5 * interquantile_range)
        upper_inner_fence = quantilearr[3].astype(float) + (1.5 * interquantile_range)
        lower_outer_fence = quantilearr[1].astype(float) - (3 * interquantile_range)
        upper_outer_fence = quantilearr[3].astype(float) + (3 * interquantile_range)
        
       

        dictionaryresult = {"name" : str(i) , "mean" : mean, "minimal" : mini , "Maximal": maxi ,
                            "Var" : vari, "Std" : tipicd, "median" : mediana, "mode" : exmode , "Upper_out_fence" : upper_outer_fence , 
                            "Lower_out_fence" : lower_outer_fence , 
                             "lower_in_fence" : lower_inner_fence , "up_in_fence" :  upper_inner_fence , "quantiles" : quantilearr,
                           "fishers skewness" : stats.skew(ex), "Pearson var cof" : stats.variation(ex) ,
                            "Kurtosis cof" : stats.kurtosis(ex) , "Std" : ex.std()}
        
        totallist.append(dictionaryresult)
        
    totallist = pd.DataFrame(totallist)
    totallist.set_index("name", inplace = True)
    totallisttrans = totallist.T
    return(totallisttrans)
        
allbasicstats(winedata)
```

## Variable study based on their relation with other variables

```python
def allbasicrelcof(dataframeargument,yvariable, typeoftarget = "cont"):
    " this function calculates for a dataframee, the relation coeficients of all features for a particular yvariable or target, \
    those coeficients are pearsons correlation, r squared,  and mutual info. Arguments that have to be specified are the dataset,\
    the name of the target variable between ' '  and finally specification if the target is continous or discrete. Anything \
    except 'cont' will consider the variable as discrete, changing the mutual info method used "
    
    import numpy as np 
    from scipy import stats
    from sklearn.feature_selection import mutual_info_regression
    from sklearn.feature_selection import mutual_info_classif
    import pandas as pd 
    
    
    yvariabledata = dataframeargument[yvariable]
    featuredata =  dataframeargument.drop(yvariable , axis = 1)
    
    
    if typeoftarget == "cont" :  
        mutualcofi = mutual_info_regression(featuredata,yvariabledata)
        
        
    else : 
        mutualcofi = mutual_info_classif(featuredata,yvariabledata)
        
    
    completelist = []
    for i in range(0,len(mutualcofi)):
        ex2 = np.array(featuredata[featuredata.columns[i]])
        ey2 = np.array(yvariabledata)
        
        pearsoncor = stats.pearsonr(ex2,ey2)
        
        dictionaryresult = {"explaining variable" : featuredata.columns[i] , "pearson correlation cof" : pearsoncor, "Rsquared" : pearsoncor[0]**2 ,
                            "mutualinfo" : mutualcofi[i] 
                            
                           }
        
        
        completelist.append(dictionaryresult)
    
    completelist = pd.DataFrame(completelist)
    completelist.set_index("explaining variable" , inplace = True)
    completelisttrans = completelist.T
    
    print("for dependent variable " + yvariable)
    return(completelisttrans)
        
allbasicrelcof(winedata, "color")
```
